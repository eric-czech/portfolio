```python
import pandas as pd
import numpy as np
from scipy.optimize import minimize
```

```python
# Specify the x points over which the true curve will be estimated
x = np.linspace(0, 100, 1000)
n = len(x)

# Specify the true curve over the above x points
p1 = 1.8
f1 = 1/25.
a1 = 500
f2 = 1/8.
a2 = 100
y = x ** p1 + a1 * np.sin(2*np.pi * x * f1) + a2 * np.sin(2*np.pi * x * f2) + np.random.normal(scale=25, size=n)

# Plot and double-check accuracy
pd.Series(y, index=x).plot()
```

```python
# Function to run minimization and return optimial fit as well as data 
# frame containing predicted and original curve data
def get_fit(x, y, curve, score_func, x0, callback=None, options=None, **kwargs):
    
    def objective(b):
        return score_func(y, curve(x, b))
    def default_callback(b):
        print 'Current parameter guess(es): {}'.format(b)
        
    cb = callback if callback else default_callback
    op = options if options else dict(disp=True)
    
    fit = minimize(objective, x0, callback=cb, options=op, **kwargs) 
    data = pd.DataFrame({'actual': y, 'predicted': curve(x, fit['x'])}, index=x)
    return fit, data
```

```python
# Specify a function that will estimate the curve given x points and parameter 
# vector of size M where M is the number of parameters to be estimated
def curve(x, b):
    res = x ** b[0] + b[1] * np.sin(2*np.pi * x * b[2])
    return res

# Specify scoring function to minimize (as 'score_func')
from sklearn.metrics import mean_squared_error
score_func = mean_squared_error

# Set the initial guess for parameters of curve as array with length M
x0 = [.5, 400, .05]

# Set boundaries for each parameter as list of tuples (also of length M)
bounds = [(-10, 10), (0, 1000), (0, 1)]
```

```python
tfit, tdata = get_fit(x, y, curve, score_func, x0, bounds=bounds)
print tfit
tdata.plot(figsize=(12,10))
```

Apparently, it's very difficult to make any of the solver methods actually fit the real curve well.  The best way to move forward without wasting a ton of time guessing parameters is to subtract the estimate so far and model the remainder, later combining both simpler models into a more complicated final estimate:

```python
dy = pd.Series(tdata.actual - tdata.predicted)
dy.plot()
```

```python
dcurve = lambda x,b: b[0] * x + b[1] * np.sin(2 * np.pi * x * b[2]) + b[3] * np.sin(2 * np.pi * x * b[4])
x0 = [0, 300, 1/25., 100, 1/10.]
bounds = [(0,.5), (300,500), (1/30.,1/25.), (50,150), (1/15., 1/5.)]
dfit, ddata = get_fit(x, dy, dcurve, score_func, x0, bounds=bounds)
print dfit
ddata.plot()
```

```python
def yhat(x):
    return curve(x, tfit['x']) + dcurve(x, dfit['x'])
pd.DataFrame({'predicted': yhat(x), 'actual': y}, index=x).plot()
```


## Minimal Example

```python
# This example will estimate a function with true value x ^ 2

curve = lambda x,b: x ** b[0] # Used in objective function to get curve evaluated with 'score_func'
x = np.linspace(0, 100, 100)  # X domain for curve
y = x ** 2                    # True curve to estimate
bbound = [(0,5)]              # Bounds for single parameter estimate
b0 = [.5]                     # Initial guess for parameter
score_func = mean_squared_error # Using MSE to evaluate parameter estimates

fit, data = get_fit(x, y, curve, score_func, b0, bounds=bbound)
print fit
data.plot()
```
