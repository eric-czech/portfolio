```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from pbn import operations as ops
from pbn import conversions as convert
%matplotlib inline
```

```python
import matplotlib.image as matimg
img_rgb = matimg.imread('/Users/eczech/Downloads/afrimov_pallette.png')
img_rgb.shape
```

```python
plt.imshow(img_rgb)
```

```python
#img_lab = convert.rgb_to_lab(img_rgb)
#df_lab = ops.unravel(img_lab)
df_lab = ops.unravel(img_rgb)
df_lab.head()
```

```python
df_counts = df_lab\
    .applymap(lambda x: np.round(x, 2))\
    .groupby(['l', 'a', 'b']).size()\
    .reset_index(name='count')\
    .sort('count', ascending=False).head(100)
df_counts.head()
```

```python
import matplotlib

colors = df_counts[['l', 'a', 'b']].head(10)
#rgb_converter = lambda c: convert._lab_to_rgb(c)
rgb_converter = lambda c: c

fig = plt.figure()
fig.set_size_inches((16,1))
ax = fig.add_subplot(111)
width, height = 1000, 200
interval = int(float(width) / len(colors))
i = -1
for _, v in colors.iterrows():
    i += 1
    offset = i * interval
    c = rgb_converter(v)
    #print(c.tolist())
    #color = [ 1 if c > 1 else c for c in color ]
    rect = matplotlib.patches.Rectangle((offset,0), interval, height, color=c)
    ax.add_patch(rect)
plt.xlim([0, width])
plt.ylim([0, height])
```
